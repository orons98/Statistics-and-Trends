```python
#importing packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

```

```python
df = pd.read_csv('Dataset_spine.csv')
df.head()
```

```python
column_name_mapping = {
    'Col1': 'pelvic incidence',
    'Col2': 'pelvic tilt',
    'Col3': 'lumbar lordosis angle',
    'Col4': 'sacral slope',
    'Col5': 'pelvic radius',
    'Col6': 'degree spondylolisthesis',
    'Col7': 'pelvic slope',
    'Col8': 'Direct tilt',
    'Col9': 'thoracic slope',
    'Col10': 'cervical tilt',
    'Col11': 'sacrum angle',
    'Col12': 'scoliosis slope',
    'Class_att': 'class' 
}

# df will have respective column names as in dataset
df.columns = [column_name_mapping.get(col, col) for col in df.columns]
numeric_df = df.drop(["class", "Unnamed: 13"], axis = 1)

print(numeric_df)
```

```python
df_describe = numeric_df.describe()
print(str(df_describe))
```

```python
#calculating the mean
df_mean = numeric_df.mean()
print(str(df_mean))
```

```python
#calculating the median
df_median = numeric_df.median()
print(df_median)
```

```python
#calculating the standard deviation
df_std = numeric_df.std()
print(df_std)
```

```python
df_skew = numeric_df.skew()
print(df_skew)
```

```python
df_kurtosis = numeric_df.kurtosis()
print(df_kurtosis)
```

```python
def scatter_plot_tilt_spondylolisthesis(df):
    """
    Plots a scatter plot to visualize the relationship between 'Pelvic Tilt' and 'Degree Spondylolisthesis'.
    
    Parameters:
    - df (pandas.DataFrame): A DataFrame containing at least 'Pelvic Tilt' and 'Degree Spondylolisthesis' columns.

    """
    
    fig, ax = plt.subplots(dpi=144)
    ax.scatter(df['pelvic tilt'], df['degree spondylolisthesis'])
    ax.set_title('Pelvic Tilt vs Degree Spondylolisthesis')
    ax.set_xlabel('Pelvic Tilt')
    ax.set_ylabel('Degree Spondylolisthesis')
    plt.show()
```

```python
scatter_plot_tilt_spondylolisthesis(df)
```

```python
def categorical_graph_quality(df):
    """
    Plots a bar chart showing the distribution of categories within the 'class' column of the given DataFrame.

    Parameters:
    - df (pandas.DataFrame): A DataFrame containing a 'class' column with categorical data.
     """

    fig, ax = plt.subplots(dpi=144)
    category_counts = df['class'].value_counts()
    ax.bar(category_counts.index, category_counts.values)
    ax.set_title('Class Distribution')
    ax.set_xlabel('Class')
    ax.set_ylabel('Count')
    plt.show()
```

```python
categorical_graph_quality(df)
```

```python
def plot_heatmap(df):
    """
    Plots a heatmap of the correlation matrix for numeric columns in the given DataFrame.

    Parameters:
    - df (pandas.DataFrame): The DataFrame from which to calculate the correlation matrix and plot the heatmap.
    
     """
        
    numeric_df = df.select_dtypes(include=[np.number])
    corr = numeric_df.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap='RdBu')
    plt.title('Heatmap of Correlation Matrix')
    plt.show()
```

```python
plot_heatmap(df)
```
